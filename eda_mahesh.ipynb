```python
import seaborn as sns
import matplotlib.pyplot as plt

#Load the built in mpg dataset
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
sns.histplot(mpg["mpg"], kde=True, bins =20)
plt.title("Distribution of miles per gallon among vehicles")
plt.xlabel("Miles per gallon")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot (data=mpg, x="mpg", y="horsepower")
plt.title("mpg vs horsepower")
plt.xlabel("miles per gallon")
plt.ylabel("horsepower")
plt.show
```

```python
sns.boxplot(data=mpg, x="origin", y ="mpg")
plt.title("miles per gallon by country of origin")
plt.xlabel("origin")
plt.ylabel("mpg")
plt.show
```

```python
sns.violinplot(data=mpg, x="origin", y="mpg")
plt.title("miles per gallon by country of origin")
plt.xlabel("origin")
plt.ylabel("mpg")
plt.show()
```

```python
sns.heatmap(mpg.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#set a Seaborn style
sns.set(style="whitegrid")

#Bar chart: Average mpg by model year
plt.figure(figsize=(8, 5))
sns.barplot(x="model_year", y="mpg", data=mpg)
plt.title("Average miles per gallon by year model")
plt.ylabel("miles per gallon")
plt.xlabel("model year")
plt.tight_layout()
plt.show()
```
